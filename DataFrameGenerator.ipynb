### 1) Для генерации данных в random устанавливаем seed = "АДВИ_2023", длина каждого списка N = 1000
### 2) index колонка: создаем список из N последовательных целых элементов от 1 до 1000 включительно, далее, с помощью random.shuffle перемешиваем его

```python
import random
import numpy as np
import pandas as pd

seed = "АДВИ_2023"
b = random.seed(a=seed)

N = 1000
index = np.arange(1, 1001)
random.shuffle(index)
```

### 3) groups колонка: создаем список из N элементов, состоящий из значений из списка ["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"], используем функцию random.choices, которая возвращает список элементов длины k, выбранных из последовательности population с перестановкой элементов

```python
lst = np.array(["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"])
groups = random.choices(lst, k=1000)
```

### 4) uniform колонка: создаем список из N элементов, состоящий из целых значение из равномерного распределения [50, 100] 

```python
uniform = []
for i in range(1, 1001):
    a = random.randint(50, 100)
    uniform.append(a)
```

### 5) gauss_1, gauss_2, gauss_3 колонки: Создаем 3 списка из N элементов каждый, заполненные значениями из нормальных распределений с параметрами:
* μ=0,   σ=1  -> третья колонка gauss_1
* μ=0,   σ=12 -> третья колонка gauss_2
* μ=50,  σ=9  -> третья колонка gauss_3

```python
gauss_1 = []
for i in range(1000):
    c = random.gauss(0, 1)
    gauss_1.append(c)

gauss_2 = []
for i in range(1000):
    d = random.gauss(0, 12)
    gauss_2.append(d)

gauss_3 = []
for i in range(1000):
    e = random.gauss(50, 9)
    gauss_3.append(e)
```

### 6) Создаем pandas DataFrame, где (1) будет индексом, (2)-(5) колонки, названия колонок ["groups", "uniform", "gauss_1", "gauss_2", "gauss_3"]

```python
Table = {
    "indexx": (index),
    "groups": (groups),
    "uniform": (uniform),
    "gauss_1": (gauss_1),
    "gauss_2": (gauss_2),
    "gauss_3": (gauss_3),
}
df = pd.DataFrame(Table)
df
```

### 7) Для колонок gauss_1 и gauss_2 заменяем значения на None:
* Для gauss_1, если индекс значение делится на 121 без остатка
* Для gauss_2, если дробная часть больше 0.95

```python
df.loc[(df.index % 121 == 0), "gauss_1"] = None
df
```

```python
df.loc[(df["gauss_2"] % 1 >= 0.95), "gauss_2"] = None
df.describe(include="all")
```

### 8) Для пропущенных значений gauss_1 и gauss_2:
* Для gauss_2 None заменить на среднее по колонке
* Удалить строчки из таблицы, где gauss_1 == None

```python
df["gauss_2"] = df["gauss_2"].fillna(df["gauss_2"].mean())
df.describe(include="all")
```
